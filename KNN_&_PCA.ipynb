# **----THEORETICAL QUESTION------**

1. What is K-Nearest Neighbors (KNN) and how does it work?
- Ans:- K-Nearest Neighbors (KNN) is a supervised learning algorithm that classifies new instances based on the majority vote of its k nearest neighbors. It works by calculating the distance between the new instance and all other instances in the training data, selecting the k nearest neighbors, and assigning the most common class label among them.


2. What is the difference between KNN classification and KNN regression?
-  Ans:- KNN classification is used for classification problems, where the target variable is categorical, and the algorithm predicts the class label of a new instance based on the majority vote of its k nearest neighbors. KNN regression is used for regression problems, where the target variable is continuous, and the algorithm predicts the value of a new instance based on the average value of its k nearest neighbors.


3. What is the role of the distance metric in KNN?
-  Ans:- The distance metric is used to calculate the similarity between instances in the feature space. Common distance metrics used in KNN include Euclidean distance, Manhattan distance, and Minkowski distance. The choice of distance metric can significantly impact the performance of the KNN algorithm.


4. What is the curse of dimensionality in KNN?
- Ans:- The curse of dimensionality refers to the phenomenon where the performance of KNN degrades as the number of features (dimensions) increases. This is because the distance between instances becomes less meaningful in high-dimensional space, making it more difficult for KNN to identify the nearest neighbors.


5. How can we choose the best value of k in KNN?
-  Ans:-The best value of k can be chosen using cross-validation, where different values of k are evaluated on a validation set, and the value that results in the best performance is selected.


6. What are KD Tree and Ball Tree in KNN?
- Ans:- KD Tree and Ball Tree are data structures used to efficiently search for nearest neighbors in KNN. KD Tree is a binary tree that partitions the data into smaller regions, while Ball Tree is a tree-like data structure that groups instances into balls based on their distances.


7. When should you use KD Tree vs. Ball Tree?
-  Ans:- KD Tree is suitable for low-dimensional data, while Ball Tree is more efficient for high-dimensional data. The choice of data structure depends on the specific characteristics of the dataset.

8.What are the disadvantages of KNN?
- Ans:- KNN has several disadvantages, including:

  - Computational complexity: KNN can be slow for large datasets.
  - Sensitivity to noise: KNN is sensitive to noisy data.
  - Curse of dimensionality: KNN performance degrades in high-dimensional space.


9. How does feature scaling affect KNN?
- Ans:- Feature scaling is essential in KNN, as the algorithm is sensitive to the scale of the features. If the features have different scales, the distance metric may be dominated by the features with larger scales, leading to poor performance.


10. What is PCA (Principal Component Analysis)?
-  Ans:- Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms the original features into a new set of orthogonal features, called principal components, that capture the most variance in the data. PCA is used to reduce the dimensionality of the data, improve the performance of machine learning algorithms, and visualize high-dimensional data.

Here are the answers:












11. How does PCA work?
- Ans:- Principal Component Analysis (PCA) works by transforming the original features of a dataset into a new set of orthogonal features, called principal components, that capture the most variance in the data. The steps involved in PCA are:

  1. Standardization: The data is standardized to have zero mean and unit variance.
  2. Covariance Matrix Calculation: The covariance matrix of the standardized data is calculated.
  3. Eigenvalue and Eigenvector Calculation: The eigenvalues and eigenvectors of the covariance matrix are calculated.
  4. Component Selection: The top k eigenvectors corresponding to the largest eigenvalues are selected.
  5. Transformation: The original data is transformed into the new feature space using the selected eigenvectors.


12. What is the geometric intuition behind PCA?
- Ans:- The geometric intuition behind PCA is that it finds the directions of maximum variance in the data and projects the data onto these directions. The first principal component captures the most variance, the second principal component captures the second most variance, and so on.


13. What is the difference between Feature Selection and Feature Extraction?
-  Ans:- Feature Selection involves selecting a subset of the most relevant features from the original feature set, while Feature Extraction involves transforming the original features into a new set of features that capture the most important information.


14. What are eigenvalues and eigenvectors in PCA?
- Ans:- Eigenvalues represent the amount of variance explained by each principal component, while eigenvectors represent the directions of the principal components.


15. How do you decide the number of components to keep in PCA?
- Ans:- The number of components to keep in PCA can be decided based on the amount of variance explained by each component, the scree plot, or cross-validation.


16. Can PCA be used for classification?
- Ans:- PCA can be used as a preprocessing step for classification, but it is not a classification algorithm itself. PCA can help improve the performance of classification algorithms by reducing the dimensionality of the data.


17. What are the limitations of PCA?
- Ans:- The limitations of PCA include:

  - PCA assumes that the data is linearly correlated.
  - PCA is sensitive to outliers.
  - PCA may not work well for high-dimensional data with non-linear relationships.


18. How do KNN and PCA complement each other?
- Ans:- KNN and PCA can complement each other by using PCA to reduce the dimensionality of the data and then applying KNN to the reduced data.


19. How does KNN handle missing values in a dataset?
- Ans:- KNN can handle missing values by imputing them using the mean or median of the nearest neighbors, or by using a distance metric that ignores missing values.


20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?
-  Ans:- The key differences between PCA and LDA are:

  - PCA is an unsupervised technique that focuses on capturing the variance in the data, while LDA is a supervised technique that focuses on finding the linear discriminants that best separate the classes.
  - PCA is used for dimensionality reduction and feature extraction, while LDA is used for classification and feature extraction.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, make_regression, make_classification
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
# 21. KNN Classifier on Iris dataset
iris = load_iris()
X_iris, y_iris = iris.data, iris.target
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.3, random_state=42)
knn = KNeighborsClassifier()
knn.fit(X_train_iris, y_train_iris)
print("KNN Classifier Accuracy on Iris:", knn.score(X_test_iris, y_test_iris))
```

```python
# 22. KNN Regressor on synthetic dataset
X_reg, y_reg = make_regression(n_samples=200, n_features=1, noise=0.1, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)
knn_reg = KNeighborsRegressor()
knn_reg.fit(X_train_reg, y_train_reg)
y_pred = knn_reg.predict(X_test_reg)
print("MSE of KNN Regressor:", mean_squared_error(y_test_reg, y_pred))
```

```python
# 23. KNN with different distance metrics
knn_euclidean = KNeighborsClassifier(metric='euclidean')
knn_manhattan = KNeighborsClassifier(metric='manhattan')
knn_euclidean.fit(X_train_iris, y_train_iris)
knn_manhattan.fit(X_train_iris, y_train_iris)
print("Euclidean accuracy:", accuracy_score(y_test_iris, knn_euclidean.predict(X_test_iris)))
print("Manhattan accuracy:", accuracy_score(y_test_iris, knn_manhattan.predict(X_test_iris)))
```

```python
# 24. Decision boundaries for different k
from matplotlib.colors import ListedColormap
X_vis, y_vis = X_iris[:, :2], y_iris
X_train_vis, X_test_vis, y_train_vis, y_test_vis = train_test_split(X_vis, y_vis, test_size=0.3, random_state=42)

def plot_decision_boundaries(X, y, k_values):
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
    plt.figure(figsize=(15, 5))
    for i, k in enumerate(k_values):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X, y)
        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                             np.arange(y_min, y_max, 0.1))
        Z = knn.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
        plt.subplot(1, len(k_values), i+1)
        plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
        plt.title(f"K={k}")
    plt.tight_layout()
    plt.show()

plot_decision_boundaries(X_train_vis, y_train_vis, k_values=[1, 5, 15])
```

```python
# 25. Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_iris)
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_scaled, y_iris, test_size=0.3, random_state=42)
knn_scaled = KNeighborsClassifier()
knn_scaled.fit(X_train_s, y_train_s)
print("Accuracy with scaling:", knn_scaled.score(X_test_s, y_test_s))
print("Accuracy without scaling:", knn.score(X_test_iris, y_test_iris))
```

```python
# 26. PCA on synthetic data
X_syn, _ = make_classification(n_samples=300, n_features=5, random_state=42)
pca = PCA()
pca.fit(X_syn)
print("Explained variance ratio (synthetic data):", pca.explained_variance_ratio_)
```

```python
# 27. PCA + KNN Classifier
pca_2d = PCA(n_components=2)
X_pca = pca_2d.fit_transform(X_scaled)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y_iris, test_size=0.3, random_state=42)
knn_pca = KNeighborsClassifier()
knn_pca.fit(X_train_pca, y_train_pca)
print("Accuracy with PCA:", knn_pca.score(X_test_pca, y_test_pca))
print("Accuracy without PCA:", knn_scaled.score(X_test_s, y_test_s))
```

```python
#28. Hyperparameter Tuning with GridSearchCV
param_grid = {'n_neighbors': [1, 3, 5, 7, 9], 'metric': ['euclidean', 'manhattan']}
grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train_s, y_train_s)
print("Best parameters from GridSearchCV:", grid.best_params_)
print("Best accuracy from GridSearchCV:", grid.best_score_)
```

```python
# 30. Visualize cumulative explained variance
pca_full = PCA().fit(X_scaled)
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)
plt.figure()
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine, load_iris, make_regression, make_classification
from sklearn.impute import KNNImputer
from sklearn.decomposition import PCA
from sklearn.metrics import (accuracy_score, mean_squared_error, classification_report,
                             precision_score, recall_score, f1_score, roc_auc_score, roc_curve)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
```

```python
sns.set(style="whitegrid")
```

```python
# Load datasets
iris = load_iris()
X_iris, y_iris = iris.data, iris.target
wine = load_wine()
X_wine, y_wine = wine.data, wine.target
```

```python
# Standardize data
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)
X_wine_scaled = scaler.fit_transform(X_wine)
```

```python
# 31. KNN weights comparison
X_train, X_test, y_train, y_test = train_test_split(X_iris_scaled, y_iris, test_size=0.3, random_state=42)
for weights in ['uniform', 'distance']:
    knn = KNeighborsClassifier(weights=weights)
    knn.fit(X_train, y_train)
    print(f"KNN ({weights}) accuracy:", knn.score(X_test, y_test))
```

```python
# 32. KNN Regressor with different K
X_reg, y_reg = make_regression(n_samples=200, n_features=1, noise=0.3)
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_reg, y_reg, test_size=0.3)
errors = []
for k in range(1, 20):
    knr = KNeighborsRegressor(n_neighbors=k)
    knr.fit(X_train_r, y_train_r)
    pred = knr.predict(X_test_r)
    errors.append(mean_squared_error(y_test_r, pred))
plt.figure(); plt.plot(range(1, 20), errors, marker='o')
plt.title("KNN Regressor: MSE vs. K"); plt.xlabel("K"); plt.ylabel("MSE"); plt.show()
```

```python
# 33. KNN Imputation
X_missing = X_iris.copy()
X_missing[np.random.randint(0, 150, 20), np.random.randint(0, 4, 20)] = np.nan
imputer = KNNImputer(n_neighbors=3)
X_imputed = imputer.fit_transform(X_missing)
print("Missing values after imputation:", np.isnan(X_imputed).sum())
```

```python
# 34. PCA projection visualization
pca_2d = PCA(n_components=2)
X_pca = pca_2d.fit_transform(X_iris_scaled)
plt.figure()
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_iris, cmap='viridis')
plt.title("PCA Projection - First 2 Components"); plt.xlabel("PC1"); plt.ylabel("PC2"); plt.show()
```

```python
# 35. KD Tree vs. Ball Tree
for algo in ['kd_tree', 'ball_tree']:
    knn_tree = KNeighborsClassifier(algorithm=algo)
    knn_tree.fit(X_train, y_train)
    print(f"Accuracy using {algo}:", knn_tree.score(X_test, y_test))
```

```python
#36. PCA Scree plot (high-dimensional)
X_hd, _ = make_classification(n_samples=300, n_features=50, random_state=42)
pca_hd = PCA().fit(X_hd)
plt.plot(np.cumsum(pca_hd.explained_variance_ratio_))
plt.title("Scree Plot"); plt.xlabel("Number of Components"); plt.ylabel("Cumulative Explained Variance"); plt.grid(True); plt.show()
```

```python
# 37. Precision, Recall, F1
knn_eval = KNeighborsClassifier()
knn_eval.fit(X_train, y_train)
y_pred = knn_eval.predict(X_test)
print("Precision:", precision_score(y_test, y_pred, average='macro'))
print("Recall:", recall_score(y_test, y_pred, average='macro'))
print("F1 Score:", f1_score(y_test, y_pred, average='macro'))
```

```python
# 38. PCA components vs. accuracy
for n in range(1, 5):
    X_pca_n = PCA(n_components=n).fit_transform(X_iris_scaled)
    X_tr, X_te, y_tr, y_te = train_test_split(X_pca_n, y_iris, test_size=0.3)
    acc = KNeighborsClassifier().fit(X_tr, y_tr).score(X_te, y_te)
    print(f"PCA components: {n}, Accuracy: {acc:.3f}")
```

```python
# 39. Leaf size effect
for leaf in [5, 20, 50]:
    clf = KNeighborsClassifier(leaf_size=leaf)
    clf.fit(X_train, y_train)
    print(f"Leaf size {leaf} accuracy:", clf.score(X_test, y_test))
```

```python
#43
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score

# Load and prepare data
iris = load_iris()
X, y = iris.data, iris.target
X_scaled = StandardScaler().fit_transform(X)
y_bin = label_binarize(y, classes=np.unique(y))  # Binarize for ROC-AUC

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_bin, test_size=0.3, random_state=42)

# Train KNN Classifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

# Predict probabilities
y_score = knn.predict_proba(X_test)

# If y_score is a list (per class), stack it
if isinstance(y_score, list):
    y_score = np.hstack([p[:, 1].reshape(-1, 1) for p in y_score])

# ROC-AUC Score
auc = roc_auc_score(y_test, y_score, average='macro', multi_class='ovr')
print("ROC-AUC Score (multi-class OvR):", auc)
```
